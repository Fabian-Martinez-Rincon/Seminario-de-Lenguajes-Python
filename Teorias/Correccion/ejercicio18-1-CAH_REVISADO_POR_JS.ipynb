EJERCICIO 18: dado el archivo denominado [Pokemon.csv](https://archivos.linti.unlp.edu.ar/index.php/s/l0v1FI2FlSKpVnK) que contiene datos sobre pokemones, se desea generar un archivo con nombres de los pokemones de tipo fuego (Type 1: Fire) que tienen un valor de ataque (Attack) mayor al de defensa (Defense).

URL: https://drive.google.com/file/d/19n9d-U_dWBQKS7oJLm0E3hl3HjiNgVa4/view?usp=sharing

```python
# Solución
import os
import csv
import json


ruta = os.path.dirname(os.path.realpath('.'))
ruta_completa = os.path.join(ruta,'Practica','Pokemon.csv')
archivo = open (ruta_completa, 'r')
nombres_fuego = open ('nombres_fuego.txt','w')


def IsTypeFire(pokemon: list):
    return (pokemon[3] == 'Fire')

def AttackGreaterThanDefense(pokemon: list):
    attack = pokemon[7]
    defense = pokemon[8]
    return (attack > defense)


csvreader = csv.reader(archivo, delimiter = ',')
encabezado = next(csvreader)

lista_nombres = []

for pokemon in csvreader:
    if IsTypeFire(pokemon) and AttackGreaterThanDefense(pokemon):
        name = pokemon[2]
        lista_nombres.append(name)

json.dump(lista_nombres,nombres_fuego)


archivo.close()
nombres_fuego.close()
```

```python
# Solución alternativa
import os
import csv
import json

def is_fire_and_offensive(pokemon: list[str]) -> bool:
    print(pokemon[6], pokemon[7], pokemon[2])
    return (pokemon[2] == 'Fire' and int(pokemon[6]) > int(pokemon[7]))

cwd = os.path.dirname(os.path.realpath('.'))
pokemon_csv_path = os.path.join(cwd,'Pokemon.csv')

with (open(pokemon_csv_path, 'r',encoding='utf-8') as csv_file,
      open('offensive_fire_pokemons.json','w',encoding='utf-8') as json_file):
    
    pkm_reader = csv.reader(csv_file, delimiter = ',')
    _ = next(pkm_reader)
    filtered_names = [pokemon[1] for pokemon in pkm_reader if is_fire_and_offensive(pokemon)]
    json.dump(filtered_names, json_file)
```
